**Question 2:** Check for missing values in all columns and describe your approach for
handling them.
(Include your Python code and output in the code box below.)

```python
# Question 2: Check for missing values in all columns
# and describe approach for handling them

import pandas as pd

# Read dataset
df = pd.read_csv("BikeDetails.csv")

# Check missing values
print("Missing values in each column:\n")
print(df.isnull().sum())

# Total missing values
print("\nTotal missing values in dataset:", df.isnull().sum().sum())

# Show data types also (important for deciding how to handle missing data)
print("\nData Types Information:\n")
print(df.info())
```

***Sample Output (example dataset):***

```python
Missing values in each column:

name              0
selling_price     0
year              0
seller_type       2
owner             5
dtype: int64

Total missing values in dataset: 7

Data Types Information:
<class 'pandas.core.frame.DataFrame'>
RangeIndex: 301 entries, 0 to 300
Data columns (total 5 columns):
 #   Column         Non-Null Count  Dtype
---  ------         --------------  -----
 0   name           301 non-null    object
 1   selling_price  301 non-null    int64
 2   year           301 non-null    int64
 3   seller_type    299 non-null    object
 4   owner          296 non-null    object
dtypes: int64(2), object(3)
```

Approach for Handling Missing Values

For numeric columns (e.g., selling_price, year):
→ Replace missing values with median (better than mean when outliers exist).

For categorical columns (e.g., seller_type, owner):
→ Replace missing values with the mode (most frequent value).
→ If a column has too many missing values (say >40%), consider dropping the column.

***Example Handling Code:***

```python
# Fill missing numeric values with median
df['selling_price'].fillna(df['selling_price'].median(), inplace=True)
df['year'].fillna(df['year'].median(), inplace=True)

# Fill missing categorical values with mode
df['seller_type'].fillna(df['seller_type'].mode()[0], inplace=True)
df['owner'].fillna(df['owner'].mode()[0], inplace=True)
```

**Question 3:** Plot the distribution of selling prices using a histogram and describe the
overall trend.
(Include your Python code and output in the code box below.)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read dataset
df = pd.read_csv("BikeDetails.csv")

# Plot histogram
plt.figure(figsize=(8,6))
plt.hist(df['selling_price'], bins=30, edgecolor='black')
plt.title("Distribution of Bike Selling Prices")
plt.xlabel("Selling Price")
plt.ylabel("Frequency")
plt.show()
```

  Expected Output (Histogram)

The histogram will show how many bikes fall into different selling price ranges.

Typically in such datasets:

Most bikes are low to mid price range (like ₹20,000 – ₹80,000).

Very few are in high price range (> ₹1,50,000).

So the graph is usually right-skewed (long tail on the right).

  Description of Overall Trend

The distribution is not uniform; it is right-skewed.

Majority of bikes are sold in the lower price range (budget bikes).

A smaller number of bikes are sold at very high prices (premium bikes).

This suggests that most owners sell cheaper bikes and only a few transactions involve costly or luxury bikes.

**Question 4:** Create a bar plot to visualize the average selling price for each seller_type
and write one observation.
(Include your Python code and output in the code box below.)

```python
# Question 4: Bar plot for average selling price by seller_type

import pandas as pd
import matplotlib.pyplot as plt

# Read dataset
df = pd.read_csv("BikeDetails.csv")

# Group by seller_type and calculate mean selling_price
avg_price_by_seller = df.groupby('seller_type')['selling_price'].mean()

# Plot bar chart
plt.figure(figsize=(8,6))
avg_price_by_seller.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Average Selling Price by Seller Type")
plt.xlabel("Seller Type")
plt.ylabel("Average Selling Price")
plt.xticks(rotation=0)
plt.show()
```

  **Expected Output:**

A bar plot with categories on X-axis (Dealer, Individual, Trustmark Dealer) and their average selling prices on Y-axis.

  **Example Observation:**

The average selling price for Dealers is usually higher compared to Individuals.

This suggests that Dealers tend to sell newer or premium bikes, while Individuals mostly sell older or budget bikes.

***Question 6:*** Use the IQR method to detect and remove outliers from the km_driven
column. Show before-and-after summary statistics.
(Include your Python code and output in the code box below.)

```python
import pandas as pd

# Read dataset
df = pd.read_csv("BikeDetails.csv")

# ---- Before removing outliers ----
print("Summary statistics BEFORE removing outliers:\n")
print(df['km_driven'].describe())

# IQR method
Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter data
df_no_outliers = df[(df['km_driven'] >= lower_bound) & (df['km_driven'] <= upper_bound)]

# ---- After removing outliers ----
print("\nSummary statistics AFTER removing outliers:\n")
print(df_no_outliers['km_driven'].describe())

# Rows before and after
print("\nNumber of rows before:", len(df))
print("Number of rows after :", len(df_no_outliers))
```

Expected Output (Example)

Before removing outliers:

```python
count     301.000000
mean    45000.123456
std     32000.567890
min      500.000000
25%    18000.000000
50%    35000.000000
75%    60000.000000
max   800000.000000
Name: km_driven, dtype: float64
```

##### **After removing outliers:**

```python
count    285.000000
mean   42000.987654
std    21000.456789
min     5000.000000
25%    17000.000000
50%    34000.000000
75%    58000.000000
max    120000.000000
Name: km_driven, dtype: float64
```

**Rows before: 301**

**Rows after: 285**

**Observation:**

Before removal, the maximum km_driven value was extremely high (like 800,000 km), which is unrealistic for bikes.

After applying the IQR method, such extreme outliers are removed, and the data becomes cleaner and more reliable.

**Question 9:** Generate a heatmap of the correlation matrix for all numeric columns.

```python
# Question 9: Heatmap of correlation matrix for numeric columns

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read dataset
df = pd.read_csv("BikeDetails.csv")

# Compute correlation matrix
corr_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap of Numeric Columns")
plt.show()
```
